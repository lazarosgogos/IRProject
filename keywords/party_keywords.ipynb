```python
import sys

sys.path.append('..')

import importcsv

print('Importing csv file...........')
small_dataset = importcsv.opencsv('../processedspeechfinal.csv')  
```

```python
#Create a dataframe to store all the results of the keywords finder for every party at each year
import pandas as pd

party_keywords_all_years = pd.DataFrame(columns= ['party', 'year', 'keywords'])
```

```python
#For each party and each year extract the keywords and save them as string to the above df
from word_frequency_calculator import filter_dataframe
from k_decomposition import getkeywords

party_names = small_dataset['political_party'].unique()

for party in party_names:
    print(party)
    for year in small_dataset.loc[small_dataset['political_party'] == party, 'year'].unique():
        print(year)
        filterd_dataframe = filter_dataframe(small_dataset, party = party, year = year)
        keywords = getkeywords(filterd_dataframe)   
        print(keywords[:20])
        small_dataset.loc[(small_dataset['political_party'] == party) &
                          (small_dataset['year'] == year), 
                          'keywords'] =  ', '.join(keywords[:20]) 
        #new_data = {'party': party, 'year': year, 'keywords':', '.join(keywords[:20])}
        #df_temp = pd.DataFrame([new_data])        
        #party_keywords_all_years = party_keywords_all_years.add(df_temp, axis=0)
        #party_keywords_all_years = pd.concat([party_keywords_all_years, df_temp], ignore_index=True, axis=0)
```

```python
df_party_years = small_dataset.drop(columns=['member_name','parliamentary_period', 'sitting_date', 'parliamentary_session', 'parliamentary_sitting', 'government','member_region',  'roles', 'member_gender', 'speech', 'processed_speeches'])
```

```python
grouped = df_party_years.groupby('keywords')
```

```python
grouped = grouped.agg({
    'political_party': lambda x: ' '.join(x.unique()),
    'year': lambda x: sum(x.unique())}).reset_index()
```

```python
grouped = grouped.sort_values(by='year')
```
