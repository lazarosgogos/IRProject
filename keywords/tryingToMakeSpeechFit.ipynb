```python
import sys

sys.path.append('..')

import importcsv

print('Importing csv file...........')
small_dataset = importcsv.opencsv('../processedspeech2.csv')  
```

```python
# remove numbers except 1453 and numbers between 1800-2039 (years)

import re

def remove_numbers(text):
    pattern = re.compile(r'\b(?!(1453 | 18[0-9]{2} | 19[0-9]{2} | 20[0-3][0-9])\b)\d+\b')
    text = re.sub(pattern, '', text)    
    return text

small_dataset['processed_speeches'] = small_dataset['processed_speeches'].apply(remove_numbers)
```

```python
# function to extract the year from a date
from datetime import datetime
def extr_year(date):
    dateobjext = datetime.strptime(date, '%d/%m/%Y')
    return dateobjext.year
```

```python
#Apply the function to column 'sitting date' and store it in a new column 'year'
small_dataset['year'] = small_dataset['sitting_date'].apply(extr_year)
```

```python
df = small_dataset['speech']

# Calculate the length of each row in bytes
row_lengths_bytes = df.apply(lambda row: sys.getsizeof(row))

# Define the threshold (128 KB in bytes)
threshold_bytes = 128 * 1024

# Use a boolean mask to filter rows based on the condition
filtered_rows = df[row_lengths_bytes > threshold_bytes]

# Display the rows that meet the condition
filtered_rows
```

```python
small_dataset = small_dataset.drop(index=filtered_rows.index.to_list()).reset_index(drop=True)
```

```python
importcsv.writetodisk(small_dataset, 'speeches__.csv')
```

```python
#For each party and each year extract the keywords and save them as string to the above df
from word_frequency_calculator import filter_dataframe
from k_decomposition import getkeywords

member_name_list = small_dataset['member_name'].unique()


for member in member_name_list:
    
    for year in small_dataset.loc[small_dataset['member_name'] == member, 'year'].unique():
        print(year)
        filterd_dataframe = filter_dataframe(small_dataset, year = year, member=member)
        keywords = getkeywords(filterd_dataframe)   
        print(keywords[:20])
        small_dataset.loc[(small_dataset['member_name'] == member) &
                          (small_dataset['year'] == year), 
                          'keywords'] =  ', '.join(keywords[:20]) 
```

```python
importcsv.writetodisk(small_dataset, 'member_keywords_all_years.csv')
```
