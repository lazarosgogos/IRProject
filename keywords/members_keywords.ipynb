```python
import sys

sys.path.append('..')

import importcsv

print('Importing csv file...........')
small_dataset = importcsv.opencsv('../processedspeechfinal.csv')  
```

```python
#For each party and each year extract the keywords and save them as string to the above df
from word_frequency_calculator import filter_dataframe
from k_decomposition import getkeywords

member_name_list = small_dataset['member_name'].unique()


for member in member_name_list:
    
    for year in small_dataset.loc[small_dataset['member_name'] == member, 'year'].unique():
        print(year)
        filterd_dataframe = filter_dataframe(small_dataset, year = year, member=member)
        keywords = getkeywords(filterd_dataframe)   
        print(keywords[:20])
        small_dataset.loc[(small_dataset['member_name'] == member) &
                          (small_dataset['year'] == year), 
                          'keywords'] =  ', '.join(keywords[:20]) 
```

```python
importcsv.writetodisk(small_dataset, 'member_keywords_all_yearsfinal.csv')
```

```python
df_member_years = small_dataset.drop(columns=['political_party','parliamentary_period', 'sitting_date', 'parliamentary_session', 'parliamentary_sitting', 'government','member_region',  'roles', 'member_gender', 'speech', 'processed_speeches'])
```

```python
grouped = df_member_years.groupby('keywords')
```

```python
grouped = grouped.agg({
    'member_name': lambda x: ' '.join(x.unique()),
    'year': lambda x: sum(x.unique())}).reset_index()
```

```python
grouped = grouped.sort_values(by='year')
grouped = grouped.drop(0)
```
