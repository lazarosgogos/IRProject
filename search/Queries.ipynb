## Import csv file and create a list with all the speeches

```python
import sys

sys.path.append('..')
import importcsv



print('Importing csv file...........')
small_dataset = importcsv.opencsv('../precessedspeechfinal.csv')  
```

```python
small_data = small_dataset['processed_speeches'].tolist()
```

# Queries algorithm

Εισαγωγή του /των ορου /-ων αναζήτησης

```python
sys.path.append('../preprocessing')
from wordtools import process_string

query = 'κληρονομιά'

query = process_string(query)

query
```

Εμφανιση της δομης του small data

```python
small_data[1]
```

Εισαγωγή του query ως τελευταία γραμμή του πίνακα με τα δεδομένα

```python
temp_small_data = small_data[:]

temp_small_data.append(query)

temp_small_data[-1] # show appended query
```

Χρειάζεται περίπου 1,5 λεπτό για να υπολογίσει τον πίνακα tfidf matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(temp_small_data)
```

Ομοιότητα συνημιτόνου

```python
from sklearn.metrics.pairwise import cosine_similarity

ddsim_matrix = cosine_similarity(tfidf_matrix[len(temp_small_data)-1,:], tfidf_matrix)
```

Πολλές ομιλίες έχουν ομοιότητα = 0. 
Διατήρηση μόνο των ομιλιών που έχουν ομοιότητα > 0.

```python
similar = {}
for i in range(len(temp_small_data)-1):
    if ddsim_matrix[0,i] > 0:
        similar[i] = ddsim_matrix[0,i]
```

```python
# Let's see what's up
similar
```

Επιλογή των top k πιο όμοιων speeches με το query

```python
from heapq import nlargest

k = 80
top_k_items = nlargest(k, similar.items(), key=lambda x: x[1])
```

```python
top_k_items
```

Εμφάνιση των καλύτερων ομιλιών με βάση το ερώτημα

```python
for i,list in enumerate(top_k_items):    
    print('Similar speech:', i)
    print('speech:',small_dataset.at[list[0],'speech'])
    #print(dataset3.at[speechstr,'processed1'])
    print('member_name:',small_dataset.at[list[0],'member_name'])
    print('political_party:',small_dataset.at[list[0],'political_party'])
    print('sitting_date:',small_dataset.at[list[0],'sitting_date'])
    if i == 6:
        break
```
