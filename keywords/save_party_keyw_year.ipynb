```python
import sys

sys.path.append('..')

import importcsv

print('Importing csv file...........')
small_dataset = importcsv.opencsv('../processedspeech.csv')  
```

#### Δημιουργία νέας στήλης με τα έτη.

```python
# function to extract the year from a date
from datetime import datetime
def extr_year(date):
    dateobjext = datetime.strptime(date, '%d/%m/%Y')
    return dateobjext.year
```

```python
#Apply the function to column 'sitting date' and store it in a new column 'year'
small_dataset['year'] = small_dataset['sitting_date'].apply(extr_year)
```

```python
#Drop rows with empty cells
small_dataset = small_dataset.dropna()
```

```python
# Change party names in speeches in order to be identified as keywords
from wordtools import change_party_names

small_dataset['processed_speeches'] = change_party_names(small_dataset)
```

```python
#Create a dataframe to store all the results of the keywords finder for every party at each year
import pandas as pd

party_keywords_all_years = pd.DataFrame(columns= ['party', 'year', 'keywords'])
```
